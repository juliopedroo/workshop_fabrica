Algumas conclusões:

3. Observando gráfico de vendas, podemos observar que boa parte está concentrada até mais ou menos 6mil euros
4. Quantos aos clientes, podemos observar que temos em média 600 clientes/dia
5. Quanto ao atributo "open" podemos observar que na maioria do tempos as lojas estavam abertas

7. Cerca de 18-20% dos dias com as lojas abertas, tínhamos um "feriado escolar"

- Plot

```python
grafico = df2.groupby('Date')[['Sales']].mean().plot(figsize=(10,5), marker = 'o', color = 'b')
grafico.set_title('Vendas por Dia')
```

Ficou muito estranho, muitos dados, não dá pra extrair nada. O que podemos fazer?

- Criando novas colunas para Dia - Mês - Ano 

Para isso vamos utilizar o recurso DatatimeIndex do Pandas na nossa coluna Date

```python
# Processo para Ano
df2['Year'] = pd.DatetimeIndex(df2['Date']).year
df2.head()
```

```python
#Mesmo processo para criarmos atributor de dia e mês
df2['Month'] = pd.DatetimeIndex(df2['Date']).month
df2['Day'] = pd.DatetimeIndex(df2['Date']).day
```

```python
df2.head()
```

```python
grafico2 = df2.groupby('Month')[['Sales']].mean().plot(figsize=(10,5), marker = 'v', color = 'y')
grafico2.set_title('Média de Vendas por Mês')
```

Podemos observar que os melhores meses das lojas são Novembro e Dezembro, com uma média de vendas maior (dezembro muito superior)

Vamos ver o comportamento da quantidade de Clientes por mês, o que vocês acham que vai acontecer? Os gráficos serão pareceidos? Diferentes? Vejamos

```python
grafico2 = df2.groupby('Month')[['Customers']].mean().plot(figsize=(10,5), marker = 'p', color = 'r')
grafico2.set_title('Média de Clientes por Mês')
```

Vamos comparar os 2 gráficos? 

- Subplot

```python
df2.groupby('Month')[['Sales']].mean().plot(figsize=(10,5), marker = 'v', color = 'y')
df2.groupby('Month')[['Customers']].mean().plot(figsize=(10,5), marker = 'p', color = 'r')
```

O comportamento se repete, com os meses de novembro e dezembro se destacando, com dezembro bem mais na frente

Será que algum dia do mês destaca-se em vendas? Ou temos as vendas distribuidas de forma linear por todo o mês?

```python
grafico3 = df2.groupby('Day')[['Sales']].mean().plot(figsize=(10,5), marker = '.', color = 'r')
grafico3.set_title('Média de Vendas por Dia')
```

Alguns picos no começo, meio e fim do mês, alguem chuta o que pode justificar isso?
